#Feature selection as folows.

```python
from sklearn.model_selection import train_test_split
```

```python
feature=df.drop(['target','flower_name'],axis='columns')
feature
```

#target

```python
target=df['target']
target
```

#train_test_split file 

```python
x_train ,x_test ,y_train ,y_test =train_test_split(feature,target,test_size=0.20)
```

using svm classification

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

using Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier,export_graphviz
from sklearn import tree
from IPython.display import SVG
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score
```

```python
classf = DecisionTreeClassifier(criterion='entropy', max_leaf_nodes=8, max_depth=3 ,random_state=0)
classf.fit(x_train,y_train)
```

```python
y_predicted = classf.predict(x_test)
y_predicted 
```

```python
accuracy_score(y_predicted , y_test)
```

using KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
 
knn = KNeighborsClassifier(n_neighbors = 11)
 
knn.fit(x_train, y_train)
pred = knn.predict(x_test)
```

```python
accuracy_score(pred , y_test)
```
